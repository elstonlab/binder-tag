## Quick visualizations for parameter sets found using the evolutionary algorithm when using the validation data set 
### Validation data set was simulated with known, ground-truth parameters

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns

# from matplotlib import rcParams
# rcParams['pdf.fonttype'] = 42
# rcParams['ps.fonttype'] = 42
# rcParams['font.family'] = 'arial'
# # rcParams['font.sans-serif'] = ['Tahoma']
# rcParams.update({'font.size': 16})
```

```python
runs = np.array(pickle.load(open('BT_simSF_0311_Fixedk3_CorrectN.pickled','rb')))
# gt_mse = pickle.load(open('GT_MSEs.pickled','rb'))
```

```python
np.shape(runs)
```

```python
plt.figure(figsize=(3,7))
sns.stripplot(y=runs[:,0,-1],orient='vertical')
sns.boxplot(y=runs[:,0,-1],color='lightgray',showfliers=False)

plt.title('EA Final Fit MSE N =')
plt.ylabel('MSE From Simulated Data')

plt.tight_layout()
# plt.savefig('EASimDataSFFinalMSE.png',dpi=150)
```

```python
runs[:,1,-1]
```

```python
original = [6,7,2,1,13,0.2,0.1] # ground truth parameters
names = ['k1','k2','k4','k6','k7','k8','f1']

amin = np.argmin(runs[:,0,:])
best = runs[:,1,:].flatten()[amin]
print(best)
```

```python
pmed = []
plt.figure(figsize=(10,10))
for i in range(7):
    plt.subplot(2,4,i+1)
    p = [params[i] for params in runs[:,1,-1]]
    pmed.append(np.median(p))
    plt.title('Median = %.2f' %np.median(p))
    sns.stripplot(y=p,orient='vertical')
#     plt.scatter(x=1,y=original[i],c='yellow')
    sns.boxplot(y=p,color='lightgray',showfliers=0)
    plt.scatter(x=0,y=original[i],c='red',zorder=10)
    plt.scatter(x=0,y=best[i],c='magenta',zorder=10)
    plt.ylabel('Parameter %s' %names[i])
    plt.tight_layout()
    print(names[i] + ' Original: %.2g, Best: %.2g Median: %.2g' %(original[i],best[i],np.median(p)))
    
```

```python
import os
import numpy as np
import pickle
```

```python
files = os.listdir('BT_simSF_0120_ceil100')
files = [file for file in files if '.txt' not in file ]
```

```python
runs_expand = []

for file in files:
    curr = pickle.load(open('BT_simSF_0120_ceil100/'+file,'rb'))
    runs_expand.append(curr)
```

```python
runs_expand = np.array(runs_expand)
```

```python
original = [6.49,  6.63,  1.79,  2.39, 1.30,12.5,  0.17,  0.13]
names = ['k1','k2','k3','k4','k6','k7','k8','f1']

amin = np.argmin(runs_expand[:,0,:])
best = runs_expand[:,1,:].flatten()[amin]
print(best)
```

```python
pmed = []
plt.figure(figsize=(10,10))
for i in range(8):
    plt.subplot(2,4,i+1)
    p = [params[i] for params in runs_expand[:,1,-1]]
    pmed.append(np.median(p))
    plt.title('Median = %.2f' %np.median(p))
    sns.stripplot(y=p,orient='vertical')
#     plt.scatter(x=1,y=original[i],c='yellow')
    sns.boxplot(y=p,color='lightgray',showfliers=0)
    plt.scatter(x=0,y=original[i],c='red',zorder=10)
    plt.scatter(x=0,y=best[i],c='magenta',zorder=10)
    plt.ylabel('Parameter %s' %names[i])
    if i !=7:
        plt.ylim([0,100])
    else:
        plt.ylim([0,1])
    plt.tight_layout()
    print(names[i] + ' Original: %.2g, Best: %.2g Median: %.2g' %(original[i],best[i],np.median(p)))
    
```

```python
pmed = []
plt.figure(figsize=(10,10))
for i in range(8):
    plt.subplot(2,4,i+1)
    p = [params[i] for params in runs_expand[:,1,-1]]
    pmed.append(np.median(p))
    plt.title('Median = %.2f' %np.median(p))
    plt.yscale('log')
    sns.stripplot(y=p,orient='vertical')
#     plt.scatter(x=1,y=original[i],c='yellow')
    sns.boxplot(y=p,color='lightgray',showfliers=0)
    plt.scatter(x=0,y=original[i],c='red',zorder=10)
    plt.scatter(x=0,y=best[i],c='magenta',zorder=10)
    plt.ylabel('Parameter %s' %names[i])
    plt.tight_layout()
```
